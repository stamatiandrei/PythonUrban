##### ***Задача. Написать класс City***

Написать класс `City` с 2 атрибутами экземплара класса и 1 методом.  
Атрибуты:
- **name** - название города
- **population** - население города

Методы:
- **get_city_info** - Метод должен возвращать строку следующего формата:  
`Город - "название текущего города". Население = "население текущего города".`.  

Инициализировать 2 экземпляра класса.  
Распечать атрибут **name** от первого экземпляра.  
От второго экземпляра распечатать результат метода **get_city_info**. 

```python
class City:
    def __init__(self, name: str, population: int):
      if not isinstance(name, str):
        raise TypeError('Введите строковое значение')
      if not isinstance(population, int):
        raise TypeError('Введите числовое значение')
      if population < 0:
        raise ValueError('Введите положительное число')

      self.name = name
      self.population = population

    def get_city_info(self) -> str:
      return f"Город - {self.name}. Население = {self.population}"
```

```python
moscow_city = City(
    'Москва'
    ,4500
)

spb_city = City(
    'Санкт-Петербург'
    ,6500
)
```

```python
print(moscow_city.name, spb_city.get_city_info(), sep="\n")
```
